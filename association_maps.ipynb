```python
from nilearn import image, datasets, plotting, input_data, maskers
import matplotlib.pyplot as plt
import numpy as np
```

```python
association_img = image.load_img('data/alzheimer_association-test.nii.gz')
atlas = datasets.fetch_atlas_aal()
masker = input_data.NiftiLabelsMasker(labels_img=atlas['maps'], standardize=False)
parcellated_map = masker.fit_transform(association_img)
```

```python
plt.hist(parcellated_map[0], bins=100)
plt.plot()
```

```python
plotting.plot_glass_brain(association_img, colorbar=True, plot_abs=False, title='Alzheimer association map')
```

```python
sorted_indices = np.argsort(-parcellated_map[0])
print('Top regions: ')
for i in sorted_indices[:10]:
    print(atlas['labels'][i])
```

## Autism
